```python
#Convert into lower text
clean_text_1=[]
def toLower(data):
    for words in raw_text:
        clean_text_1.append(str.lower(words))
```

```python
toLower(raw_text)
```

```python
clean_text_1
```

```python
#Stage 2 -- tokenize
clear_text_2=[]

from nltk.tokenize import sent_tokenize,word_tokenize

import nltk
nltk.download('punkt')
```

```python
sent_tok=[]
for sent in clean_text_1:
    sent=sent_tokenize(sent)
    sent_tok.append(sent)
```

```python
sent_tok
```

```python
# Word tokenize
clean_text_2=[ word_tokenize(i) for i in clean_text_1]
```

```python
clean_text_2
```

```python
#Removes Punctuations

import re

clean_text_3=[]

for words in clean_text_2:
    clean=[]
    for w in words:
        res=re.sub(r'[^\w\s]' , "", w)
        if res!="":
            clean.append(res)
    clean_text_3.append(clean)
```

```python
clean_text_3
```

```python
#Stage 4
#Removing Stop Words -- repeated words

import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
clean_text_4=[]

for words in clean_text_3:
    w=[]
    for word in words:
        if not word in stopwords.words('english'):
            w.append(word)
    clean_text_4.append(w)
```

```python
clean_text_4
```

```python
#Stage5 Stemming
from nltk.stem.porter import PorterStemmer
```

```python
port=PorterStemmer()
```

```python
clean_text5=[]

for words in clean_text_4:
    w=[]
    for word in words:
        w.append(word)
    clean_text5.append(w)
```

```python
clean_text5
```

```python
#Lemmatization

from nltk.stem.wordnet import WordNetLemmatizer
```

```python
wnet=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lem=[]

for words in clean_text_4:
    w=[]
    for word in words:
        w.append(wnet.lemmatize(word))
    lem.append(" ".join(w))
```

```python
lem
```

```python
df['PreProcessed_News']=lem
```
